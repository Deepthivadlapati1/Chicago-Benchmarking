```python
df = df.drop(['Water Use (kGal)','District Steam Use (kBtu)','District Chilled Water Use (kBtu)','All Other Fuel Use (kBtu)','Row_ID','ID','Latitude','Longitude','ZIP Code','Address'],axis=1)
```

```python
df.shape
```

```python
num_cols = df.select_dtypes(include=['number']).columns
cat_cols = df.select_dtypes(include=['object', 'category']).columns

#creating new dataframes for numerical and categorical columns
df_numerical = df[num_cols]
df_categorical = df[cat_cols]

#displaying the result for numerical and categorical columns
print("Numerical Columns are:")
print(df_numerical.head())

print("------------------------------------------------------------------------------------")

print("Categorical Columns are:")
print(df_categorical.head())
```

```python
print(df_numerical.shape)
print(df_categorical.shape)
```

```python
#filling the missing values for numerical columns with mean
for num_column in num_cols:
    df[num_column].fillna(df[num_column].mean(), inplace=True)

#fill the missing values for categorical columns with mode
for cat_column in cat_cols:
    df[cat_column].fillna(df[cat_column].mode()[0], inplace=True)
```

```python
#displaying the combined dataframe 
print("\nFinal combined DataFrame:")
print(df.head())
```

```python
#importing necessary libraries for building algorithms
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
X = df_encoded.drop('Reporting Status', axis=1)  
y = df_encoded['Reporting Status']  

#Spliting the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
```

```python
from sklearn.naive_bayes import GaussianNB

# Initialize and fit Naive Bayes classifier
nb_classifier = GaussianNB()
nb_classifier.fit(X_train, y_train)

# Make predictions
y_pred = nb_classifier.predict(X_test)

# Evaluate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f'Naive Bayes Accuracy: {accuracy:.2f}')
```

```python
#Initializing the model and training the model
logistic_model = LogisticRegression(max_iter=1000)
logistic_model.fit(X_train, y_train)

#Making the predictions
y_pred = logistic_model.predict(X_test)

#Evaluating the model
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred)

print(f'Accuracy: {accuracy:.2f}')
print('Confusion Matrix:')
print(conf_matrix)
print('Classification Report:')
print(class_report)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score, classification_report

#Initializing the model and training the model
gb_clf = GradientBoostingClassifier(n_estimators=100, learning_rate=0.1, random_state=42)
gb_clf.fit(X_train, y_train)

#Making the predictions
y_pred_gb = gb_clf.predict(X_test)

#Evaluating the model
gb_cv_scores = cross_val_score(gb_clf, X, y, cv=5)
print(f"Gradient Boosting Cross-Validation Accuracy: {gb_cv_scores.mean():.2f}")
```

```python
from sklearn.tree import DecisionTreeClassifier

#Initializing the model and training the model
dt_classifier = DecisionTreeClassifier()
dt_classifier.fit(X_train, y_train)

#Making the predictions
y_pred = dt_classifier.predict(X_test)

#Evaluating the model
dt_cv_scores = cross_val_score(dt_classifier, X, y, cv=5)
print(f"Decision Tree Cross-Validation Accuracy: {dt_cv_scores.mean():.2f}")
```

1. Which properties have the least energy performance based on EUI Site Electricity Use Intensity and EUI Source?

```python
import matplotlib.pyplot as plt
import seaborn as sns

#Sorting the properties by Site EUI and Source EUI to get the least efficient ones
top_least_efficient = df.sort_values(by=['Site EUI (kBtu/sq ft)', 'Source EUI (kBtu/sq ft)'], ascending=False).head(10)

#Bar plot
plt.figure(figsize=(12, 6))
sns.barplot(x='Property Name', y='Site EUI (kBtu/sq ft)', data=top_least_efficient, color='b', label='Site EUI')
sns.barplot(x='Property Name', y='Source EUI (kBtu/sq ft)', data=top_least_efficient, color='r', label='Source EUI')

plt.xticks(rotation=45, ha="right")
plt.title('Top 10 Properties with the Least Energy Performance')
plt.ylabel('Energy Use Intensity (kBtu/sq ft)')
plt.legend()
plt.tight_layout()
plt.show()
```

2. Are there any trends which may be observed in energy performance of properties depending on their type, location, or age?

```python
#Box plot for Site EUI based on property type
plt.figure(figsize=(15, 6))
sns.boxplot(x='Primary Property Type', y='Site EUI (kBtu/sq ft)', data=df)
plt.xticks(rotation=45, ha="right")
plt.title('Site EUI by Property Type')
plt.tight_layout()
plt.show()
```

```python
#Scatter plot to observe trends in energy performance based on year built
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Year Built', y='Site EUI (kBtu/sq ft)', data=df, hue='Primary Property Type', palette='coolwarm', size='Gross Floor Area - Buildings (sq ft)', sizes=(20, 200))
plt.title('Energy Performance (Site EUI) vs. Year Built')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
#plt.tight_layout()
plt.show()
```

3. How does the intensity of greenhouse gas emission differ across property types which buildings contribute the most?

```python
#Grouping by property type and calculating mean GHG Intensity
ghg_by_property_type = df.groupby('Primary Property Type').agg({'GHG Intensity (kg CO2e/sq ft)': 'mean'}).reset_index()

#Bar plot 
plt.figure(figsize=(12, 6))
sns.barplot(x='Primary Property Type', y='GHG Intensity (kg CO2e/sq ft)', data=ghg_by_property_type, palette='viridis')
plt.xticks(rotation=45, ha="right")
plt.title('Average GHG Intensity by Property Type')
plt.tight_layout()
plt.show()
```
